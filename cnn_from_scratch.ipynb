## <img src="static_images/icons/noun-advance-2109145.png" alt=">" style="width:50px"/> Step 5: Putting everything together

Allright, good job getting here! Now it's time to see if all our hard work pays off. In the following cell we will train your model and validate it against the validation set.

Let's start by defining a few hyperparameters. Feel free to experiment with different values and try to optimize your model:

````python
batch_size = 64         # Keep the batch size, or experiment with 32 or 128
valid_size = 0.2        # Fraction of the training data to reserve for validation
num_epochs = 75         # Increase number of epochs
num_classes = 50        # Number of classes (fixed)
dropout = 0.6           # Increase dropout to prevent overfitting
learning_rate = 0.0001  # Reduce learning rate for more gradual training
opt = 'adam'            # You can still try SGD with momentum if needed
weight_decay = 0.001    # Add some weight decay to prevent overfitting
````

````python
from src.data import get_data_loaders
from src.train import optimize
from src.optimization import get_optimizer, get_loss
from src.model import MyModel
import torch

# Get data loaders
data_loaders = get_data_loaders(batch_size=batch_size, valid_size=valid_size, num_workers=-1)

# Initialize the model
model = MyModel(num_classes=num_classes, dropout=dropout)

# Load best model checkpoint
try:
    model.load_state_dict(torch.load('./checkpoints/best_val_loss.pt'))
    print("Loaded model checkpoint successfully.")
except FileNotFoundError:
    print("No checkpoint found. Training from scratch.")

# Get optimizer using the parameters defined above
optimizer = get_optimizer(
    model=model, 
    optimizer=opt, 
    learning_rate=learning_rate, 
    weight_decay=weight_decay
)

# Get loss function
loss = get_loss()

# Train model (or continue training from checkpoint)
optimize(
    data_loaders,
    model,
    optimizer,
    loss,
    n_epochs=num_epochs,
    save_path="checkpoints/best_val_loss.pt",
    interactive_tracking=True
)
````

Allright, now we are ready to export our model using our Predictor class:

````python
from src.predictor import Predictor
from src.helpers import compute_mean_and_std
from src.model import MyModel
from src.data import get_data_loaders
import torch

# Load the data loaders with a batch size of 1 for inference
data_loaders = get_data_loaders(batch_size=1)

# Get the class names from the training data in the data loaders
class_names = data_loaders["train"].dataset.classes

# Instantiate the model and move it to the CPU (as we don't need GPU for inference)
model = MyModel(num_classes=50, dropout=0.5).cpu()

# Load the best model weights (make sure they were saved earlier during training)
# We load the weights on the CPU using the 'map_location' argument
model.load_state_dict(torch.load('checkpoints/best_val_loss.pt', map_location='cpu'))

# Get the dataset's mean and standard deviation for normalization purposes
mean, std = compute_mean_and_std()

# Wrap the model using the Predictor class, along with the computed mean, std, and class names
predictor = Predictor(model, class_names, mean, std).cpu()

# Export the Predictor model using torch.jit.script to convert it into TorchScript
scripted_predictor = torch.jit.script(predictor)

# Save the scripted model to a file for future use
scripted_predictor.save("checkpoints/original_exported.pt")

print("Model successfully exported to 'checkpoints/original_exported.pt'")
````
